## 2. Сборка графов
Для облегчения ручной сборки создадим 2 вспомогательных функции:

- Ф-ция принимает список длин линий, возвращает несвязанный граф, соответствующий линиям:

```python
def create_lines(lens):
    pre_intersect = {}
    for n, length in zip([i + 1 for i in range(len(lens))], lens):
        v = [set() for i in range(length)]
        for i in range(length - 1):
            v[i+1].add((n, i+1))
            v[i].add((n, i+2))
        pre_intersect.update(dict(zip([(n, i + 1) for i in range(length)], v)))
    return pre_intersect
```

- Ф-ция принимает список пересечений линий и добавляет их в граф:

```python
def add_intersect(lens, inter):
    for a, b in inter:
        for i in lens[b]:
            lens[i].discard(b)
            lens[i].add(a)
        lens[a].update(lens.pop(b))
    return lens
```

## 3. Проверка на изоморфизм

### Первый способ:

```python
from networkx.algorithms import isomorphism

GM = isomorphism.GraphMatcher(G_peter,G_cairo_cut)
GM.is_isomorphic()
```

### Второй способ:
- Поскольку графы метро имеют специфический вид с большим количеством "хвостов" и "перемычек" (подграфов из вершин степени 2 в виде путей), будет разумным попробовать построить сжатую форму таких графов, где хвосты и перемычки заменяются одной вершиной, которой присваивается вес, равный количеству замещённых вершин. Прочим вершинам присвоим вес 1.
- Т.к. граф и его сжатая версия имеют взаимооднозначное соответствие, оригинальные графы будут изоморфны тогда и только тогда, когда изоморфны их сжатые версии с учетом весов вершин.

Следующая ф-ция преобразует графы в соответствии с описанным выше:

```python
def compress(graph):
    G = graph.copy()
    count = 0
    for v in G.nodes:
        if G.degree(v) != 2:
            G.node[v]['weight'] = 1
        else:
            p1, p2 = G.neighbors(v)
            if G.degree(p1) > 2 and G.degree(p2) > 2:
                G.node[v]['weight'] = 1
            else:
                G.node[v]['weight'] = 0
                count += 1

    for i in range(count):
        for v in G:
            if G.degree(v) == 2 and G.node[v]['weight'] != 1:
                G.node[v]['weight'] = max(1, G.node[v]['weight'])
                p1, p2 = G.neighbors(v)
                if G.degree(p1) == 1:
                    G.node[v]['weight'] += G.node[p1]['weight']
                    G.remove_node(p1)
                elif G.degree(p2) == 1:
                    G.node[v]['weight'] += G.node[p2]['weight']
                    G.remove_node(p2)
                elif G.degree(p1) == 2:
                    G.node[v]['weight'] += max(G.node[p1]['weight'], 1)
                    neib = list(G.neighbors(p1))
                    neib.remove(v)
                    G.add_edge(v,neib[0])
                    G.remove_node(p1)
                elif G.degree(p2) == 2:
                    G.node[v]['weight'] += max(G.node[p2]['weight'], 1)
                    neib = list(G.neighbors(p2))
                    neib.remove(v)
                    G.add_edge(v,neib[0])
                    G.remove_node(p2)
                else:
                    continue
                break
    return G
```

Для изоморфизма с учетом весов необходимым условием становится не просто совпадение упорядоченных наборов степеней вершин, а совпадение упорядоченных наборов пар вес/степень вершины!

```python
def weight_degree(G):
    m = {}
    for v in G.nodes:
        temp = m.pop((G.node[v]['weight'], G.degree(v)), [])
        temp.append(v)
        m[(G.node[v]['weight'], G.degree(v))] = temp
    return m
```

```python
wd_peter = weight_degree(G_comp_peter)
wd_peter
```

```python
wd_cairo = weight_degree(G_comp_cairo)
wd_cairo
```

- Для наших графов эти наборы очевидно не совпадают.
- Чтобы проверить дальнейшую часть алгоритма, соорудим из графа для Каира альтернативный переименовыванием вершин.
- Такой граф очевидно должен быть изоморфен изначальному

```python
cairo_alt = nx.convert_node_labels_to_integers(G_cairo_cut, ordering='increasing degree')
```

```python
cairo_alt.nodes()
```

```python
G_comp_cairo_alt = compress(cairo_alt)
wd_cairo_alt = weight_degree(G_comp_cairo_alt)
wd_cairo_alt
```

- Теперь наборы совпадают!
- Но чтобы убедиться в изоморфизме, нам нужно зафиксировать порядок типов вершин и самих вершин в одном графе, зафиксировать в том же порядке типы вершин второго графа, и последовательным перебором комбинаций в рамках одного типа найти такую, что соответствующая матрица смежности второго графа в точности станет соответствовать матрице смежности первого.
- Данный способ по сути является умным брутфорсом (в худшем случае мы переберём для нашего конкретного графа $4 \cdot 2 \cdot 2 \cdot 3 \cdot 2 \cdot 2 = 192$ комбинации вместо потенциальных $61!$

Фиксируем порядо одного графа:

```python
order = []
for key in sorted(wd_cairo):
    order.extend(sorted(wd_cairo[key]))
print(order)
```

Для второго фиксируем порядок типов, но не порядок вершин:

```python
order_alt = []
for key in sorted(wd_cairo_alt):
    order_alt.append(wd_cairo_alt[key])
print(order_alt)
```

Следующая функция рекурсивно перебирает возможные порядки и сравнивает матрицы смежности:

```python
from itertools import permutations

def checker(order_fix, order_alt, a_g, H):
    if len(order_alt) == 0:
        a_h = nx.adjacency_matrix(H, order_fix)
        if (a_g - a_h).nnz == 0:
            return True
        else:
            return False
    temp = list(permutations(order_alt[0]))
    for t in temp:
        fix = order_fix.copy()
        fix.extend(t)
        if checker(fix, order_alt[1:], a_g, H):
            return True
    return False
         
```

Проверим на наших графах Каира:

```python
a_g = nx.adjacency_matrix(G_comp_cairo, order)
checker([], order_alt, a_g, G_comp_cairo_alt)
```

Верно!  

Теперь соберём всё в один метод:

```python
from itertools import permutations

def are_isomorphic(G, H):
    """Check whether two graphs G and H are isomorphic.
    
    Note: This function is brute force and very slow.
    
    args:
        G: a networkx Graph
        H: a networkx Graph
    
    returns:
    	True if G and H are isomorphic.
    	False if G and H are not isomorphic.
    """
    n = len(G.nodes())
    m = len(H.nodes())
    if n != m:
        return False
    if sorted(dict(G.degree()).values()) != sorted(dict(H.degree()).values()):
        return False

    G = compress(G)
    H = compress(H)
    n = len(G.nodes())
    m = len(H.nodes())
    if n != m:
        return False

    wd_G = weight_degree(G)
    wd_H = weight_degree(H)
    
    if sorted(wd_G) != sorted(wd_H):
        return False
    
    order = []
    for key in sorted(wd_G):
        order.extend(sorted(wd_G[key]))

    order_alt = []
    for key in sorted(wd_H):
        order_alt.append(wd_H[key])

    a_g = nx.adjacency_matrix(G, order)
    return checker([], order_alt, a_g, H)

    return False
```

```python
are_isomorphic(G_cairo_cut, G_peter)
```
